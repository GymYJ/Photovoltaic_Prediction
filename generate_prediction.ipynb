```
for date in tempGNT.index:
  for hour in tempGNT.columns:
    t[date+' '+hour]=tempGNT[hour][date]
```

```
temp=pd.DataFrame(t.values(), index=t.keys())
```

```
print(temp)
```

```
temp.columns=['발전량']
temp.index.names=['일자']
temp.head()
```

```
def min_max_normalize(lst):
  normalize = []

  for value in lst:
    normalized_num=(value - min(lst))/(max(lst)-min(lst))
    normalize.append(normalized_num)

  return normalize
```

```
wtData['발전량']=0
wtData.head()
```

```
wtData['시간']=0
for date in wtData.index:
  temp=date[11:13]
  if temp[1]==":":temp=temp[0]
  wtData['시간']=int(temp)
```

```
for date1 in wtData.index:
  for date2 in temp.index:
    if date1==date2:
      wtData['발전량'][date1]=temp['발전량'][date2]
```

```
wtData=wtData.fillna(-1)
wtData.head()
```

```
feature=["기온(°C)","강수량(mm)","습도(%)","현지기압(hPa)","지면온도(°C)","시간"]
gen=["발전량"]
T_data=wtData[feature]
T_label=wtData["발전량"]
```

```
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors=900,weights="distance")
```

```
print(T_label)
```

```
regressor.fit(T_data,T_label)
```
